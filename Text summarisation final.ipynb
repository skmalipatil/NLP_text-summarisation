```python
from nltk.tokenize import WhitespaceTokenizer
from nltk.tokenize import word_tokenize
from nltk.tokenize import MWETokenizer
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
from nltk.stem import WordNetLemmatizer
from nltk.probability import FreqDist
from collections import Counter
from nltk.tokenize import sent_tokenize
import numpy as np
from sklearn.pipeline import make_pipeline 
```

```python
def read_file(file_name):
    file = open(file_name, encoding = 'utf8')
    all_text = file.read()
    return(all_text)


# tokenisation of the text
# removing of the stop words
#lammatisation of the words
def preprocessing(all_text):
    #tokenisation
    words = WhitespaceTokenizer().tokenize(all_text)
    #removing of stop words
    all_words = [word for word in words if word.lower() not in stop_words]
    #Lammatisation of the words
    lammatised_word = [WordNetLemmatizer().lemmatize(each, pos ='v') for each in all_words]
    return(lammatised_word)


def words_scoring(lammatised_word):
    freq = Counter(lammatised_word)
    return(freq)


def score_sen(text):
    sentences = sent_tokenize(text)
    score_of_sentences = []
    for sentence in sentences:
        words = [word for word in WhitespaceTokenizer().tokenize(sentence)]
        count = 0
        freq = Counter(words)
        for word in words:
            if word in freq.keys():
                count = count+freq[word]
        score_of_sentences.append(count)
    print(len(score_of_sentences))
    return(score_of_sentences)


#printing of the sentences
def text_summary(text):
    score = score_sen(text)
    sentences = sent_tokenize(text)
    summary_text = []
    for i in range(len(score)):
        if score[i] > np.ceil(2*np.mean(score)):
            summary_text.append(sentences[i])
    print(len(summary_text))
    return('. '.join(summary_text))
```

```python
#name = input()
def text_summarisation(file_name):
    read_text = read_file(file_name)
    lamattized_word = preprocessing(read_text)
    freq = words_scoring(lamattized_word)
    return(text_summary(read_text))

#text_summarisation(name)
```
